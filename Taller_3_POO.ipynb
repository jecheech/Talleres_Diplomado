### Los programas de TV que puede elegir son:

- Código 1: "Avengers", por un precio de 10000 COP
- Código 2: "Serie Dark", por un precio de 15000 COP
- Código 3: "InterEstelar", por un precio de 8000 COP
- Código 4: "Piratas del Caribe", por un precio de 12000 COP

### La comida que puede encargar es :

- Código 1: Restaurantes Favoritos, sin costo adicional.
- Código 2: (Menú estándar de Cinepython) Perro, por un precio de 15000 COP
- Código 3: (Menú estándar de Cinepython) Hambuerguesa, por un precio de 18000 COP
- Código 4: (Menú estándar de Cinepython) Gaseosa, por un precio de 6000 COP

```python
class Servicio_Contratar():
    def sala(self,tipo_sala,horas):
        if tipo_sala == 1:
            precio=50000
            resumen = """El tipo de sala elejida {}, por un precio de {} COP por hora, incluye la libertad de escoger las películas/series que desea ver y la posibilidad de pedir comida de sus restaurantes favoritos para consumir dentro de la sala. Incluyendo el menú estándar de Cinepython.""".format(tipo_sala, precio)
        elif tipo_sala == 2:
            precio=30000
            resumen = """El tipo de sala {}, por un precio de {} COP por hora, incluye la posibilidad de pedir comida de sus restaurantes favoritos para consumir dentro de la sala. Incluyendo el menú estándar de Cinepython. Con restricción para escoger las películas/series""".format(tipo_sala, precio)
        elif tipo_sala == 3:
            precio=20000
            resumen = """El tipo de sala {}, por un precio de {} COP por hora, no incluye la posibilidad de pedir comida de sus restaurantes favoritos para consumir dentro de la sala y tiene restricción para escoger las películas/series. Solo el menú estándar de Cinepython""".format(tipo_sala, precio)
        else:
            precio=0
            resumen = """El tipo de sala {}, no se encuentra dentro de las opciones disponibles""".format(tipo_sala)
        precio_sala = precio * horas
        return [precio_sala, resumen]
        
            
    def elije_tv(self,cod_movie):
        cod = [1,2,3,4]
        names = [["Avengers",10000],["Serie Dark",15000],["InterEstelar",8000],["Piratas del Caribe",12000]]
        dicc_pelis = {key:value for key, value in zip (cod,names)}
        precio2 = dicc_pelis[cod_movie]
        return precio2
    
    def elije_comida(self,cod_comida,cantidad):
        cod = [1,2,3,4]
        names = [["Restaurantes Favoritos",0],["Perro",15000],["Hambuerguesa",18000],["Gaseosa",6000]]
        dicc_comida = {key:value for key, value in zip (cod,names)}
        precio3 = dicc_comida[cod_comida][1]*cantidad
        return [dicc_comida[cod_comida][0], precio3]
    
    def pregunta_tv(self):
        programa = "Ninguno"
        precio_tv = 0
        tv = int(input("Indique el codigo del programa que desea ver... "))
        precio_tv = self.elije_tv(tv)[1]
        programa = self.elije_tv(tv)[0]
        return [programa, precio_tv ]

    def pregunta_comida(self):
        menu = []
        precio_comida = 0
        salir = 1
        restaurante = int(input("0 'Menú estándar de Cinepython' y 1 para 'Restaurantes Favoritos' "))
        if restaurante == 0:
            while salir!=0:
                comida = int(input("Indique el codigo de la comida que desea --Escriba 0 para salir del menu--... "))
                if comida == 0:
                    salir = 0
                    break
                else:
                    cantidad = int(input("Indique la cantidad de {} que desea... ".format(self.elije_comida(comida,1)[0])))
                    menu.append([self.elije_comida(comida,cantidad)[0], cantidad])
                precio_comida = precio_comida + self.elije_comida(comida,cantidad)[1]  
        return [precio_comida, menu]
        
        
```

```python
tipo_sala = int(input("Indique el tipo de sala..."))
tiempo = int(input("Indique por cuanto tiempo desea alquilar la sala..."))

servicio = Servicio_Contratar()
sala = servicio.sala(tipo_sala,tiempo)
resumen_servicio = sala[1]
precio_sala = sala[0]

if tipo_sala ==1:
    tv = servicio.pregunta_tv()
    comida = servicio.pregunta_comida()
         
elif tipo_sala ==2:
    comida = servicio.pregunta_comida()
            
            
total = precio_sala+tv[1]+comida[0]

resumen_servicio = """

{} 

                    A continuación el detalle de la cuenta
                    
                    1. Precio de la Sala {} total por {} horas
                    2. Precio del Programa de TV {} es {}
                    3. Precio de la comida {}
                    
                    {}
                    _______________________________________
                    
                    Total {}

                    """.format(str(resumen_servicio),precio_sala, tiempo,tv[0],tv[1],comida[0],comida[1],total)

print(resumen_servicio)
    


```
